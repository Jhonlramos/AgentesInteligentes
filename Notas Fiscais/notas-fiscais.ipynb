```python
# importing libs
import os
import zipfile
import pandas as pd
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain_experimental.agents import create_pandas_dataframe_agent
```

```python
# Caminho atual onde o Jupyter está rodando
current_dir = os.getcwd()
print(f"Caminho atual: {current_dir}")
```

```python
# Define os caminhos com base no diretório atual
zip_path = os.path.join(current_dir, "Archive.zip")
extract_dir = os.path.join(current_dir, "extracted")

# Cria a pasta de extração se não existir
os.makedirs(extract_dir, exist_ok=True)

# Descompacta o arquivo
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("Arquivos extraídos:")
print(os.listdir(extract_dir))
```

```python
# Caminhos completos para os arquivos CSV
cabecalho_path = os.path.join(extract_dir, "202401_NFs_Cabecalho.csv")
itens_path = os.path.join(extract_dir, "202401_NFs_Itens.csv")

# Leitura dos arquivos
df_cabecalho = pd.read_csv(cabecalho_path)
df_itens = pd.read_csv(itens_path)

# Verifica rapidamente se foi carregado
print("Cabecalho:")
print(df_cabecalho.shape)
print("Itens:")
print(df_itens.shape)
```

```python
class SafeDataAgent:
    def __init__(self, openai_api_key: str):
        self.llm = ChatOpenAI(
            temperature=0,
            openai_api_key=openai_api_key,
            model="gpt-4"
        )
        self.agent = None

    def carregar_dataframes(self, *dataframes):
        if len(dataframes) == 2:
            df_merged = dataframes[0].merge(dataframes[1], on="CHAVE DE ACESSO", how="left")
        else:
            df_merged = dataframes[0]

        self.agent = create_pandas_dataframe_agent(
            self.llm,
            df_merged,
            verbose=False,
            allow_dangerous_code=True
        )

    def perguntar(self, pergunta: str):
        if not self.agent:
            raise Exception("Nenhum DataFrame carregado.")
        return self.agent.run(pergunta)
```

```python
# === 🔑 Inicialização segura (via variável de ambiente) ===
openai_key = os.getenv("OPENAI_API_KEY")  # certifique-se de definir no seu sistema

# ✅ Ou, para teste rápido (NÃO RECOMENDADO para produção):
# openai_key = "sk-proj-..."

# === 🚀 Criação do agente ===
agent = SafeDataAgent(openai_api_key=openai_key)

# === 📊 Carregamento dos DataFrames já lidos (df_cabecalho, df_itens) ===
agent.carregar_dataframes(df_cabecalho, df_itens)

# === 💬 Exemplo de pergunta ===
resposta = agent.perguntar("Qual item teve o maior volume entregue?")
print("🧠 Resposta do agente:", resposta)
```
